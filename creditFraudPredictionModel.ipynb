Libraries for Project

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

Loading data into pandas dataFrame

```python
credit_data = pd.read_csv('/home/haroonwaheed-19/Downloads/DataSets for ML/credit_data.csv')
```

separating legit and fault transactions data

```python
legit = credit_data[credit_data.Class == 0]
fraud = credit_data[credit_data.Class == 1]
```

```python
print(legit,legit.shape)
print(fraud,fraud.shape)
```

```python
fraud['Amount'].describe()
```

undersampling the legit data for better performance and model training

```python
legit_sample = legit.sample(n=len(fraud),random_state=42)
```

```python
print(legit_sample.shape)
print(fraud.shape)
```

creating new dataset after undersampling and shuffling the dataset

```python
balanced_credit_data = pd.concat([legit_sample,fraud],axis=0)
balanced_credit_data = balanced_credit_data.sample(n=len(fraud),random_state=42).reset_index(drop=True)
```

```python
balanced_credit_data.shape
```

```python
balanced_credit_data.head()
```

Separating the features and labels of the balanced_credit_data

```python
x = balanced_credit_data.drop(columns='Class',axis=1)
y = balanced_credit_data['Class']
```

```python
print(x.shape)
print(y.shape)
```

Splitting the dataset into Test and Train Data

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

Model Loading and training

```python
model = LogisticRegression()
model.fit(x_train,y_train)
```

Predictions on training Data

```python
x_train_predictions = model.predict(x_train)
x_train_accuracy = accuracy_score(x_train_predictions,y_train)
print("Model Trained Accuracy Score : ",x_train_accuracy*100)
```

Prediction on testing Data

```python
x_test_predictions = model.predict(x_test)
x_test_accuracy = accuracy_score(x_test_predictions,y_test)
print("Model Test Accuracy Score : ",x_test_accuracy*100)
```

Making Function to check credit card accuracy on input data

```python
def predictCredit(input):
    input_data = np.asarray(input)
    input_data_reshaped = input_data.reshape(1,-1)
    predictions = model.predict(input_data_reshaped)
    if predictions[0] == 0:
        return "The object is a legitimate transaction"
    else:
        return "The object is a fraudulent transaction"

```
